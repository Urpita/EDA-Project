# Null value Analysis

```python
df.isnull().sum()
```

```python
for col in cat:  # Assuming 'cat' is a list of categorical column names
    mode = df[col].mode()[0]  # Get the mode (most frequent value) of the column
    df[col].fillna(mode, inplace=True)  # Fill missing values with the mode
```

```python
for col in num:  # Assuming 'num' is a list of numerical column names
    mean_value = df[col].mean()  # You can also use median() if preferred
    df[col].fillna(mean_value, inplace=True)  # Fill missing values with the mean
```

# Outliers analysis

**Fill with median**

```python
app_data=df['ApplicantIncome']
q1 = round(np.quantile(app_data, 0.25), 2)
q3 = round(np.quantile(app_data, 0.75), 2)

# Compute the IQR and the lower and upper bounds
IQR = q3 - q1
lb = q1 - 1.5 * IQR
ub = q3 + 1.5 * IQR

median = df['ApplicantIncome'].median()

# Apply np.where correctly to replace values outside the bounds with the median
df['app_data_1'] = np.where((df['ApplicantIncome'] < lb) | (df['ApplicantIncome'] > ub), 
                            median, 
                            df['ApplicantIncome'])

# Verify the result
print(df[['ApplicantIncome', 'app_data_1']].head())
```

```python
# Create a horizontal box plot for the 'p_wage_1' column
plt.boxplot(df['app_data_1'], vert=False, patch_artist=True, 
            boxprops=dict(facecolor='lightblue'))

# Add title and labels for better clarity
plt.title('Box Plot of Adjusted Application Income', fontsize=14)
plt.xlabel('Application Income', fontsize=12)

# Display the box plot
plt.show()
```
