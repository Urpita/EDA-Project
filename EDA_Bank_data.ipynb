**Outliers removing and filling with median**

```python
data=Bank_df['age']
q1=q1=round(np.quantile(data,0.25),2)
q3=round(np.quantile(data,0.75),2)
IQR=q3-q1
lb=q1-1.5*IQR
ub=q3+1.5*IQR
median=data.median()
new_data=[]
for i in data:
    if i<lb or i>ub:
        new_data.append(median)
    else:
        new_data.append(i)
Bank_df['page']=new_data
```

```python
data=Bank_df['age']
q1=round(np.quantile(data,0.25),2)
q3=round(np.quantile(data,0.75),2)
IQR=q3-q1
lb=q1-1.5*IQR
ub=q3+1.5*IQR
median=data.median()
con=(Bank_df['age']<lb) | (Bank_df['age']>ub)
true=median
false=Bank_df['age']
Bank_df['p_wage_1']=np.where(con,true,false)
```

**Fill with median**

```python
bal_data=Bank_df['balance']
q1=round(np.quantile(bal_data,0.25),2)
q3=round(np.quantile(bal_data,0.75),2)
IQR=q3-q1
lb=q1-1.5*IQR
ub=q3+1.5*IQR
median=bal_data.median()
new_data=[median if (i< lb or i>ub) else i for i in bal_data]
Bank_df['newbal']=new_data
print(Bank_df[['balance','newbal']].head(10))
```

# Encoding Methods

```python
# converting all categorical columns to Numerical columns by map function

Bank_df=pd.read_csv("bank.csv",sep=';')
for col in cat:
    d={}
    labels=Bank_df[col].unique()
    for i in range(len(labels)):
        d[labels[i]]=i
    Bank_df[col]=Bank_df[col].map(d)
Bank_df.head()
```
