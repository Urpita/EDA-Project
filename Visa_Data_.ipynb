# Handling Outliers

```python
# Extract prevailing_wage column
wage_data = visa_df['prevailing_wage']

# Calculate the 1st and 3rd quartile
q1 = round(np.quantile(wage_data, 0.25), 2)
q3 = round(np.quantile(wage_data, 0.75), 2)

# Compute the IQR
IQR = q3 - q1

# Set the lower bound and upper bound
lb = q1 - 1.5 * IQR
ub = q3 + 1.5 * IQR

# Create conditions to identify non-outliers
con1 = visa_df['prevailing_wage'] > lb
con2 = visa_df['prevailing_wage'] < ub
con3 = con1 & con2

# Filter out the non-outliers
non_outliers_data = visa_df[con3]

# Get the count of non-outliers
count = len(non_outliers_data)

# Print the result
print(f"Number of non-outliers: {count}")
print(non_outliers_data.head())  # Optional: view the filtered data
```

**Fill with median**

```python
# Extract prevailing_wage column
wage_data = visa_df['prevailing_wage']

# Calculate the 1st and 3rd quartile
q1 = round(np.quantile(wage_data, 0.25), 2)
q3 = round(np.quantile(wage_data, 0.75), 2)

# Compute the IQR and the lower and upper bounds
IQR = q3 - q1
lb = q1 - 1.5 * IQR
ub = q3 + 1.5 * IQR

# Calculate the median
median = wage_data.median()

# Replace outliers with the median using list comprehension
new_data = [median if (i < lb or i > ub) else i for i in wage_data]

# Add the new data as a column to the DataFrame
visa_df['pwage'] = new_data

# Check the results
print(visa_df[['prevailing_wage', 'pwage']].head())
```

```python


# Plot the histogram for 'pwage'
plt.hist(visa_df['pwage'], bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# Add a title and labels for clarity
plt.title('Distribution of Adjusted Prevailing Wage (pwage)', fontsize=14)
plt.xlabel('Prevailing Wage (pwage)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)

# Show the plot with tighter layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Extract prevailing_wage column
wage_data = visa_df['prevailing_wage']

# Calculate the 1st and 3rd quartile
q1 = round(np.quantile(wage_data, 0.25), 2)
q3 = round(np.quantile(wage_data, 0.75), 2)

# Compute the IQR and the lower and upper bounds
IQR = q3 - q1
lb = q1 - 1.5 * IQR
ub = q3 + 1.5 * IQR

# Calculate the median
median = wage_data.median()

# Use np.where() to replace outliers with the median
visa_df['p_wage_1'] = np.where((visa_df['prevailing_wage'] < lb) | (visa_df['prevailing_wage'] > ub),
                               median, 
                               visa_df['prevailing_wage'])

# Check the results
print(visa_df[['prevailing_wage', 'p_wage_1']].head())
```

```python
# Create a horizontal box plot for the 'p_wage_1' column
plt.boxplot(visa_df['p_wage_1'], vert=False, patch_artist=True, 
            boxprops=dict(facecolor='lightblue'))

# Add title and labels for better clarity
plt.title('Box Plot of Adjusted Prevailing Wage', fontsize=14)
plt.xlabel('Prevailing Wage (p_wage_1)', fontsize=12)

# Display the box plot
plt.show()
```
